## Задание 1: Получение однозначного числа

Пользователь вводит целое число, программа складывает все цифры числа, с полученным числом — то же самое, и так до тех пор, пока не получится однозначное число.

<br>

**Примеры:**
* `545   -> 5`
* `12345 -> 6`

````python
inp = [545, 12345]
output = [5, 6]
encoded = [0] * len(output)

def calculate_one_digit(num: int) -> int:
    result = 0
    
    while num > 9:
        result = 0
        for dig in str(num):
            result += int(dig)
        num = result

    return result

for i in range(len(inp)):
    encoded[i] = calculate_one_digit(inp[i])

assert output == encoded, 'worong result'
````

## Задание 2: Кинотеатр

Дан список списков, каждый вложенный список состоит из 1 и 0, количество вложенных списков — количество рядов. Пользователь вводит, сколько билетов ему требуется. Программа должна найти ряд, где можно приобрести нужное количество билетов (места должны быть рядом). Если таких рядов несколько, то ближайший к экрану (ближайшим считается нулевой ряд). Если таких мест нет, то вывести False.

<br>

**Примеры:**
* `[[0,1,1,0], [1,0,0,0], [0,1,0,0]], 2 -> 1`
* `[[0,1,1,0], [1,0,1,0], [1,1,0,1]], 2 -> False`

````python
inp = [ 
    [ [[0,1,1,0], [1,0,0,0], [0,1,0,0]], 2 ], 
    [ [[0,1,1,0], [1,0,1,0], [1,1,0,1]], 2 ] ]
output = [1, False]
encoded: list[int | bool] = [0] * len(output)

def find_available_row(all_sits: list[list[int]], sits_count: int) -> int | bool:
    for i in range(len(all_sits)):
        row = all_sits[i]
        available_sits = 0

        for sit in row:
            if sit == 0:
                available_sits += 1
            else:
                available_sits = 0
            
            if available_sits == sits_count:
                return i

    return False

for i in range(len(inp)):
    encoded[i] = find_available_row(*inp[i])

assert encoded == output, "wrong result"
````

## Задание 3: Алгоритм RLE

Необходимо написать упрощенную версию алгоритма RLE. 
Алгоритм RLE объединяет подряд идущие символы в коэффициент и символ.

<br>

**Примеры:**
* `asssdddsssddd -> 1a3s3d3s3d`
* `aaabbbbccccc  -> 3a4b5c`
* `abcba         -> 1a1b1c1b1a`

````python
inp = ['asssdddsssddd', 'aaabbbbccccc', 'abcba']
output = ['1a3s3d3s3d', '3a4b5c', '1a1b1c1b1a']
encoded = [''] * len(output)


def simple_rle(string: str) -> str:
    if len(string) == 0:
        return string
    elif len(string) == 1:
        return f'1{string}'
    
    count = 1
    char = string[0]
    result = ''

    for i in range(1, len(string)):
        curr_ch = string[i]
        if curr_ch == char:
            count += 1

        if curr_ch != char:
            result += f'{count}{char}'
            count = 1
            char = curr_ch

        if i == len(string) - 1:
            result += f'{count}{char}'
    
    return result
    
for i in range(len(inp)):
    encoded[i] = simple_rle(inp[i])

assert encoded == output, "wrong result"
````

## Задание 4: Шифр Цезаря

Реализуйте программу для шифрования текста с помощью шифра Цезаря.  
Шифр Цезаря — это метод шифрования, при котором каждая буква в тексте заменяется буквой, стоящей на фиксированное число позиций дальше в алфавите.

### Требования
1. Программа должна принимать на вход:
   - Строку текста для шифрования
   - Ключ шифра (число от 0 до 25)

2. Правила шифрования:
   - Шифруются только латинские буквы (a-z, A-Z)
   - Сохраняется регистр букв (заглавные остаются заглавными)
   - Пробелы, цифры и знаки препинания остаются без изменений
   - Сдвиг циклический (после 'z' идет 'a', после 'Z' идет 'A')

### Примеры

* `"Hello World", 3      -> "Khoor Zruog"`
* `"Python is great!", 5 -> "Udymts nx lwjfy!"`
* `"XYZ abc", 3          -> "ABC def"`
* `"Test 123", 1         -> "Uftu 123"`

### Алгоритм
* Для каждого символа в строке:
   - Если это латинская буква, найти её позицию в алфавите
   - Добавить к позиции значение ключа
   - Если результат больше 25, использовать остаток от деления на 26
   - Преобразовать обратно в букву, сохранив регистр
   - Если это не буква, оставить символ без изменений


````python
import string as String

inp = [('Hello World', 3), ('Python is great!', 5), ('XYZ abc', 3), ('Test 123', 1)]
output = ['Khoor Zruog', 'Udymts nx lwjfy!', 'ABC def', 'Uftu 123']
result = [''] * len(output)

alphabet = String.ascii_lowercase

MAX_IDX = 25

def cezar_encode(string: str, offset: int) -> str:
    encoded = ''

    for i in range(0, len(string)):
        curr_ch = string[i]
        is_lower = curr_ch.islower()
        curr_ch = curr_ch.lower()

        idx = alphabet.find(curr_ch)
        if idx == -1:
            encoded += curr_ch
            continue

        idx += offset
        if idx > MAX_IDX:
            idx = idx % (MAX_IDX + 1)

        encoded += alphabet[idx] if is_lower else alphabet[idx].upper()
    
    return encoded
    
for i in range(len(inp)):
    result[i] = cezar_encode(*inp[i])

assert result == output, "wrong result"
````

````python
db: dict = {}

def is_integer(s):
    try:
        int(s)
        return True
    except ValueError:
        return False

def user_input() -> bool:
    subject = input('Enter subject name: ')
    if subject == '':
        return True
    
    subject = subject.strip()

    name = input('Enter student name: ')
    name = name.strip()

    grade = 0
    while True:
        grade = input('Enter grade: ')
        if is_integer(grade):
            grade = int(grade)
            if 1 <= grade <= 5:
                break 
        print('Grade should be an integer value in 1 ..= 5 range')

    if db.get(subject) is None:
        db[subject] = []

    db[subject].append((name, grade))
    return False

def get_subject_data() -> bool:
    subject = input('Enter subject name to see students table or `q` to quit the app:')
    if subject == 'q':
        return True
    
    if db.get(subject) is None:
        print(f'Subject dosn`t exist, please enter one of these subjects: {list(db.keys())}')
        return
    
    print(subject)
    students: dict[str, list[str]] = {}
    for name, grade in db[subject]:
        if students.get(name) is None:
            students[name] = []

        students[name].append(str(grade))

    for student, grades in students.items():
        print(f'{student} {' '.join(grades)}')

    return False

if __name__ == "__main__":
    complete_input = False
    while True:
        
        if not complete_input:
            complete_input = user_input()
            continue

        should_stop = get_subject_data()
        if should_stop:
            break
````
