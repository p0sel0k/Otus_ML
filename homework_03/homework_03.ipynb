## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`



```python
import string

inp = ['otus_course', 'PythonIsTheBest']
out = ['OtusCourse', 'python_is_the_best']
res = [''] * len(out)

def snake_pascal_converter(text: str) -> str:
    if '_' in text:
        words = text.split('_')
        words = list(map(lambda w: w.capitalize(), words))
        return ''.join(words)
    
    start_idx = 0
    words = []
    for idx, ch in enumerate(text):
        if ch.isupper() and idx != 0:
            words.append(text[start_idx:idx])
            start_idx = idx
        
        if idx == len(text) - 1:
            words.append(text[start_idx:(idx+1)])

    words = list(map(lambda w: w.lower(), words))
    return '_'.join(words)        

for i in range(len(inp)):
    res[i] = snake_pascal_converter(inp[i])

assert res == out, "wrong result"
```

## Задание 2: Проверка валидности даты

Написать функцию проверяющую валидность введенной даты.

<br>

**Примеры:**
* `29.02.2000 -> True`
* `29.02.2001 -> False`
* `31.04.1962 -> False`



```python
from functools import reduce

inp = ['29.02.2000', '29.02.2001', '31.04.1962']
out = [True, False, False]
res = [False] * len(out)

days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

def is_integer(s):
    try:
        int(s)
        return True
    except ValueError:
        return False

def is_leap_year(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

def is_date_valid(day: int, month: int, year: int) -> bool:
    if year < 0:
        return False
    
    if is_leap_year(year):
        days_in_month[1] = 29
    else:
        days_in_month[1] = 28

    if not (1 <= month <= 12):
        return False
    
    if not (1 <= day <= days_in_month[month - 1]):
        return False

    return True

def parse_date(date: str) -> tuple[int, int, int]:
    assert '.' in date, 'Wrong date type please input date in this format: `dd.mm.yyyy`'
    date = date.split('.')
    assert len(date) == 3, 'Wrong date format. Date can`t constain more than 3 positions (day, month, year)'
    assert reduce(lambda a,b: a and is_integer(b), date, True), 'Date values shoud be integers'

    return tuple(map(int, date))

for i in range(len(inp)):
    res[i] = is_date_valid(*parse_date(inp[i]))

print(res)
assert res == out, "wrong result"
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
import math

inp = [17, 20, 23]
out = [True, False, True]
res = [False] * len(out)

def is_integer(s):
    try:
        int(s)
        return True
    except ValueError:
        return False


def is_number_prime(num: int):
    assert is_integer(num), "Number shoud be an integer!"
    if num <= 1:
        return False
        
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            return False
    return True


for i in range(len(inp)):
    res[i] = is_number_prime(inp[i])

print(res)
assert res == out, "wrong result"
```

## Задание 4: Учет пользователей

Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле. 

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. 

**Программа должна проверять:**
* имя и фамилия состоят только из символов и начинаются с большой буквы - если не с большой, то заменяет букву на большую;
* возраст должен быть числом от 18 до 60;
* ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным.

**Дополнительно:** написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python
db: dict = {}

def is_integer(s):
    try:
        int(s)
        return True
    except ValueError:
        return False
    
def is_letters_only(text: str) -> bool:
    for ch in text:
        if not ch.isalpha():
            return False
    return True

def user_input() -> bool:
    name = ''
    while True:
        name = input('Enter user name: ')
        if name == '':
            return True
        
        if not is_letters_only(name):
            print('Name shoud not contains any non alpabetical symbols')
            continue
        name = name.strip()
        break

    surename = ''
    while True:
        surename = input('Enter user surename: ')
        if not is_letters_only(name):
            print('Surename shoud not contains any non alpabetical symbols')
            continue
        surename = surename.strip()
        break

    age = 0
    while True:
        age = input('Enter user age: ')
        if is_integer(age):
            age = int(age)
            if 18 <= age <= 60:
                break
            print('Age should be in range from 18 to 60')
            continue
        print('Age should be an integer')
    
    id = 0
    while True:
        id = input('Enter user ID: ')

        if is_integer(id):
            id = int(id)
            if id < 0 or id > 99999999:
                print('ID should be in range from 0 to 99999999')
                continue
        else:    
            print('ID should be an integer')
            continue

        id = str(id)
        id = '0'*(8 - len(id)) + id

        if db.get(id) is None:
            db[id] = (name, surename, age)
        else:
            print('ID should be unique, please try again')
        break

    return False

def get_users_data() -> bool:
    for id, data in db.items():
        print(f'{id}: {data[0]} {data[1]} {data[2]}')

if __name__ == "__main__":
    complete_input = False
    while True:
        
        if not complete_input:
            complete_input = user_input()
            continue

        get_users_data()
        break
```
