# I. Numpy

### Импортируйте NumPy

```python
import numpy as np
```

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
zeros_array = np.zeros(10, dtype=int)
print(zeros_array)

zeros_array[4] = 1
print(zeros_array)

zeros_array = np.reshape(zeros_array, (2, 5))
print(zeros_array)
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
arr = np.arange(10, 50)[::-1]
print(arr)

result = arr[arr % 2 == 0]
print(result)
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
arr = np.arange(0, 9).reshape(3, 3)
print(arr)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
arr = np.random.random((4, 3, 2))
a_min = arr.min()
a_max = arr.max()

print(a_min, a_max)
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение.

```python
mat1 = np.random.random((6, 4))
mat2 = np.random.random((4, 3))
mat3 = mat1 @ mat2
print(mat3)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
arr = np.random.random((7, 7))
a_mean = arr.mean()
a_std = arr.std()

print(a_mean, a_std)
```

## Сделайте выводы по анализу датасета и построенным графикам. По желанию можете продолжить анализ данных и также отразить это в выводах.

```python
# Базовая информация о датасете: Данные не содержат пропусков, датасет сожержит 244 срочки и 7 признаков

# total_bill имеет "хвост" в правую сторону, то есть много небольших чеков и меньше очень больших, что видно по гистограмме.
# На графиках наглядно видна взаимосязь между суммой счета и суммой чаевых (показывает явную положительную зависимость: чем выше total_bill, тем в среднем выше tip),
# в свою очерердь на общую сумму счета больше влияют время (средний счет во время ужина выше),
# а так же дни недели (на выходных средний счет несколько выше, чем в будние дни)
# Если рассматривать значения общего чека выше медианы, то размер среднего чека у обоих полов примерно одинаковый,
# так же судя по графикам курение не окаывает влияния на total_bill или tip

sns.heatmap(tips.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation heatmap")
plt.show()
```
